```python
from datetime import date

import uuid


# here are we are making function to add new book 


book_list=[];     # here we are making list where we want to store all the books in the list as tuples

def add_new_book():

    book_information=();     # here we are making tuple of book inforation so they cannot be changed
        
    # as tuples are immutable we can not assign values using assignment operator so we assigne values like below
        
    isbn = int(input("enter the isbn number "))
        
    book_information = book_information+(isbn,);       # we are using this syntax because compilee amy not confuse between data types
        
    title = input("enter the title ");
        
    book_information = book_information+(title,);
        
    author = input("enter the author ");
        
    book_information = book_information+(author,);
        
    price = int(input("enter the price "))
        
    book_information = book_information+(price,);
        
    quantity = int(input("enter the quantity "))
        
    book_information = book_information+(quantity,);
        
    book_list.append(book_information)     #adding each new book to to the list 
        
    # print(book_information)
        
    print(book_list)

    print("book added to inventory successfully ") 


# add_new_book()


def updateTheBookQuantity():
    update_quantitybook =();
    isbn = int(input("please enter the book isbn to update the quantity "))
    # print("len",len(book_list))
    # print(book_list)
    for i in range(0,len(book_list)):
        if(book_list[i][0]==isbn):
            updated_qunatity = int(input("please enter the updated quantity of book"))
            update_quantitybook=update_quantitybook+(book_list[i][0],)
            update_quantitybook=update_quantitybook+(book_list[i][1],)
            update_quantitybook=update_quantitybook+(book_list[i][2],)
            update_quantitybook=update_quantitybook+(book_list[i][3],)
            update_quantitybook=update_quantitybook+(updated_qunatity,)
            book_list[i]=update_quantitybook;
            print("\nQuantity updated successfulyy")
            print(book_list)
            break;
        else:
            print("\nyour enter isbn number is not correct or book not found")

    # print("\n",update_quantitybook)
    # print(book_list)
# updateTheBookQuantity()


def removing_outofstock():
    for i in range(0,len(book_list)):
        if(book_list[i][4]<=0):
            book_list.pop(book_list[i][0])
        else:
            print("\nThere is no book to remove or no book out of stock ")

# removing_outofstock()


customer={}   # dictionery to store customers

sales_date={}

customedetail_list=[]

def customer_data():

    userordersbooks =[]
    confirm_order = []
    sales_data = {}
    quantity_list =[]
    quantity_after_order=()
    customer_id_and_email=()
    
    # temp_book_tuple=(book_list)


    name = input("plese enter your name ");
    email = input("please enter your email ");
    orderbook=None
    while(orderbook!="q"):
        orderbook=input("please enter the book to order and press q to exit ");
        if(orderbook!="q"):
            orderquantity=int(input("please enter the quantity "))
        quantity_list.append(orderquantity)
        userordersbooks.append(orderbook)

        #checking weather the enter book is present in invetory or not 
    # temp_user_order=(userordersbooks)
    
    # temp_booklist.update(temp_book_tuple)
    # confirm_order.update(temp_user_order);
    # print(temp_booklist.intersection(confirm_order))

    # print(temp_booklist)
    # print(confirm_order)
        

        #checkin weather user enter book is present or not in the inventory
        
    for i in range(0,len(userordersbooks)):
        for j in range (0,len(book_list)):
            if(book_list[j][1]==userordersbooks[i]):
                if(book_list[j][4]>=quantity_list[i]):              #checking weather the quantity is greater or not

                    print("order prcocessed succcessfully ")
                    #updating quantity after the order process
                    quantity_after_order=quantity_after_order+(book_list[j][0],)
                    quantity_after_order=quantity_after_order+(book_list[j][1],)
                    quantity_after_order=quantity_after_order+(book_list[j][2],)
                    quantity_after_order=quantity_after_order+(book_list[j][3],)
                    quantity_after_order=quantity_after_order+(book_list[j][4]-quantity_list[i],)
                    book_list[j]=quantity_after_order
                    confirm_order.append(userordersbooks[i])
                    sales_data[date.today()]={book_list[j][0]:quantity_list[i]}
                
                
                

    uid =uuid.uuid4()
    customer_id_and_email=customer_id_and_email+(uid,)
    customer_id_and_email=customer_id_and_email+(email,)
    customedetail_list.append(customer_id_and_email)
        
    customer[uid]={"name":name,"email":email,"orderhistory":confirm_order}

    # print(customer)
    # print(confirm_order)
    # print(book_list)
    # print(sales_data)

# customer_data()
# # customer_data()


def display_cutomer():
    email = input("plese enter the eamil of customer to get its information ")

    for i in range (0,len(customedetail_list)):
        if(email==customedetail_list[i][1]): #checking weather the cutomer exist or not
            print(customer[customedetail_list[i][0]])
            break;
        else:
            print("no customer found")


# display_cutomer()


def book_in_inventory():

    print("books in inventory are given below ")
    for i in range (0,len(book_list)):
        print(book_list[i])


def maindisplay():
    
    print("\npress 1 to add new book ")
    print("\npress 2 to update the book quantity ")
    print("\npress 3 to remove the out of stock books from inventory ")
    print("\npress 4 to order a book ")
    print("\npress 5 to see the cutomer information ")
    print("\npress 6 to see the books in inventory ")
    
    choice=None
    while(choice!=0):
        choice = int(input("please choose and press 0 to exit"))
        
        if(choice==1):
            add_new_book()
        elif(choice==2):
            updateTheBookQuantity()
        elif(choice==3):
            removing_outofstock()
        elif(choice==4):
            customer_data()
        elif(choice==5):
            display_cutomer()
        elif(choice==6):
            book_in_inventory()
        else:
            print("please enter a right choice")


maindisplay()
    
    
    

        

        
        
        
```
